### Zadanie 3.
Z tej samej ramki danych wyświetl dane wykorzystując filtry oparte o mapowanie:
* a) tylko kobiety z wykształceniem 'Master' oraz 'Bachelor' i pożyczkami pomiędzy 50000 a 70000,
* b) wszyscy mężczyźni z wykształceniem innym niż 'Master' oraz będący właścicielem lokalu (kolumna person_home_ownership = OWN),
* c) wszystkie osoby z wartością `loan_percent_income` w przedziale od 0.4 do 0.6,
* d) wszystkie wartości pożyczek, których kwota jest wielokrotnością liczby 10000.

```python
max(df.loan_amount)
```

Maksymalna pożyczka równa się 35000. Więc zadanie 3a było zmodyfikowano w następny sposób:

a) tylko kobiety z wykształceniem 'Master' oraz 'Bachelor' i pożyczkami pomiędzy 15000 a 30000


```python
df.query('person_gender == "female" & person_education == ["Master", "Bachelor"] & loan_amount > 15000 & loan_amount < 30000')
```

```python
# b) wszyscy mężczyźni z wykształceniem innym niż 'Master' oraz będący właścicielem lokalu (kolumna person_home_ownership = OWN)
df.query('person_gender == "male" & person_education != "Master" & person_home_ownership == "OWN"')
```

```python
# c) wszystkie osoby z wartością `loan_percent_income` w przedziale od 0.4 do 0.6
df.query('loan_percent_income >= 0.4 & loan_percent_income <= 0.6' )
```

```python
df[(df['loan_percent_income'] >= 0.4) & (df['loan_percent_income'] <= 0.6)]
```

```python
# d) wszystkie wartości pożyczek, których kwota jest wielokrotnością liczby 10000
df[df['loan_amount'] % 10000 == 0]
```
